```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
```

```python
im=image.load_img("cat.jpg",target_size=(200,200))
```

```python
im
```

```python
data=ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=False
    
)
```

```python
im_vector=image.img_to_array(im)
```

```python
im_vector.shape
```

```python
input_batch=im_vector.reshape((1,)+im_vector.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
i=0
for batch in data.flow(input_batch, batch_size=1,save_to_dir="aug", save_prefix="cat", save_format="jpeg"):
    i+=1
    
    if i==5:
        break
```
