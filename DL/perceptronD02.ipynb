```python
from sklearn.datasets import make_classification
import pandas as pd
import seaborn as sns
import numpy as np
X,y=make_classification(n_samples=100,n_features=2,n_classes=2,n_redundant=0,n_informative=1,random_state=42,class_sep=10,n_clusters_per_class=1,hypercube=False)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    w=np.zeros(X.shape[1])
    for _ in range(100):
       j=np.random.randint(0,X.shape[0])
       y_pred=np.dot(X[j],w)>=0
       w=w+0.01*(y[j]-y_pred)*X[j]

    return w[0],w[1:]   
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
intercept_,coef_
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input=np.array([X[:,0].min(),X[:,0].max()])
y_input=m*x_input+b
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.plot(x_input,y_input,color="red")
```
