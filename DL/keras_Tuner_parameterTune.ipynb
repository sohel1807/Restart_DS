```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("diabetes.csv")
```

```python
df.corr()["Outcome"]
```

```python
x=df.iloc[:,:-1]
y=df["Outcome"]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score
```

```python
sc=StandardScaler()
```

```python
x_scl=sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scl,y,test_size=0.2,random_state=0,stratify=y)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(32,activation="relu",input_dim=8))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100)
```

```python
import kerastuner as kt
```

```python
#Optimzier Tune
def build_model(hp):
    model=Sequential()
    model.add(Dense(32,activation="relu",input_dim=8))
    model.add(Dense(1,activation="sigmoid"))
    optimizer=hp.Choice("optimizer",values=["adam","sgd","rmsprop"])
    model.compile(loss="binary_crossentropy",optimizer=optimizer,metrics=["accuracy"]) 

    return model
```

```python
tuner=kt.RandomSearch(build_model,max_trials=10,objective="val_accuracy")
```

```python
tuner.search(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,initial_epoch=6)
```
