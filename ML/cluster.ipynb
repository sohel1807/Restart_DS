```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("student_clustering.csv")
```

```python
df
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
X=df.iloc[:,:].values
kmeans=KMeans(n_clusters=4,random_state=42)
y_kmeans=kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
X[y_kmeans==3,1]
```

```python

plt.scatter(X[y_kmeans == 0,0],X[y_kmeans == 0,1],color='blue')
plt.scatter(X[y_kmeans == 1,0],X[y_kmeans == 1,1],color='red')
plt.scatter(X[y_kmeans == 2,0],X[y_kmeans == 2,1],color='green')
plt.scatter(X[y_kmeans == 3,0],X[y_kmeans == 3,1],color='yellow')
```

```python
from sklearn.cluster import AgglomerativeClustering 
```

```python
from scipy.cluster import hierarchy as sch
```

```python
sch.dendrogram(sch.linkage(X,method='ward'))
```

```python
ac=AgglomerativeClustering(n_clusters=4,linkage='ward')
y_ac=ac.fit_predict(X)
y_ac
```

```python
plt.scatter(X[y_ac == 0,0],X[y_ac == 0,1],color='blue')
plt.scatter(X[y_ac == 1,0],X[y_ac == 1,1],color='red')
plt.scatter(X[y_ac == 2,0],X[y_ac == 2,1],color='green')
plt.scatter(X[y_ac == 3,0],X[y_ac == 3,1],color='yellow')
```

```python
from sklearn.cluster import DBSCAN
```

```python
ds=DBSCAN(eps=2,min_samples=2)
y_ds=ds.fit_predict(X)
y_ds
```

```python
plt.scatter(X[y_ds == 0,0],X[y_ds == 0,1],color='blue')
plt.scatter(X[y_ds == 1,0],X[y_ds == 1,1],color='red')
plt.scatter(X[y_ds == 2,0],X[y_ds == 2,1],color='green')
plt.scatter(X[y_ds == 3,0],X[y_ds == 3,1],color='yellow')
```
