```python
# Importing necessary Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.datasets import make_regression
from sklearn.linear_model import Ridge
```

```python
# Making sample regression data
x,y=make_regression(n_features=1,n_samples=500,noise=40,random_state=42)
```

```python
plt.scatter(x,y)
```

```python
Lr=LinearRegression()
Lr.fit(x,y)
```

```python
y_pred=Lr.predict(x)
```

```python
r2_score(y,y_pred)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color="red")
```

```python
print(Lr.coef_)
print(Lr.intercept_)
```

```python
rg=Ridge(alpha=100)
rg.fit(x,y)
```

```python
y_Rg=rg.predict(x)
```

```python
r2_score(y,y_Rg)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color="black")
plt.plot(x,y_Rg,color="red")
plt.show()
```
