```python
# Remove Punctuation
```

```python
import string,time
```

```python
string.punctuation
```

```python
exclude=string.punctuation
```

```python
def remove_punctuation(text):
    for char in exclude:
        text=text.replace(char,"")
    return text    
```

```python
remove_punctuation("Hello, World!")
```

```python
def remove_punctuation(text):
    return text.translate(str.maketrans("","",exclude))
```

```python
remove_punctuation("This is a test, to remove punctuation! Does it work? Yes, it does.")
```

```python
## Stop Words Removal
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stopwords.words("english")
```

```python
def stopwords_removal(text):
    stop_words = []
    new_text =[]
    for word in text.split(" "):
        if word.lower() in stopwords.words("english"):
            stop_words.append(word)
            new_text.append("")
        else:
            new_text.append(word)
    text = " ".join(new_text)            

    return text.strip(),stop_words        
```

```python
stopwords_removal("This is a test to remove stop words from this sentence.")
```

```python
# Stemming
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem_text(text):
    new_text=[]
    for word in text.split():
        new_text.append(ps.stem(word))
    return " ".join(new_text)
```

```python
stem_text("walking walked walks walk")
```
