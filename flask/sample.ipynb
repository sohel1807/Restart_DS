```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import pickle
```

```python

df = pd.DataFrame({
    "experience":[1,2,3,4,5],
    "salary":[30000,40000,50000,60000,70000]
})

X = df[["experience"]]
y = df["salary"]

model = LinearRegression()
model.fit(X,y)

pickle.dump(model, open("model.pkl","wb"))
```
