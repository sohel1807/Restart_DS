```python
import matplotlib.pyplot as plt
import numpy as np
```

# DAY_1
## Simple Line Chart

```python
#variable name fig to refer to a figure instance and ax to refer to an axes instance or group of axes instances.
fig=plt.figure()
ax=plt.axes()
```

```python
#simple sine graph
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```

```python
#If we want to create a single figure with multiple lines, we can simply call the plot function multiple times:
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))

# Or just Write plt.plot(x,np.sin(x))
# plt.plot(x,np.sin(y)) 
```

```python
# colour,Linestyle

plt.plot(x,np.sin(x-0),color="m") # by hexacode,RGBCMYK(Red,Green,Blue,Cyan,Magenta,Yellow,blacK)
plt.plot(x,np.sin(x-1),linestyle="dashed") # by symbol or name like dashed,dotted,solid,dashdoted etc
```

```python
# limit for x and y axis
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
# Labeling Point

plt.plot(x,np.sin(x))
plt.title("Sine Curve") # Title Of Graph
plt.xlabel("x value") # x axis label
plt.ylabel("sin(x)") # y axis label
```

```python
# Legend
plt.plot(x,np.sin(x),"-.m",label="sin(x)")
plt.plot(x,np.cos(x),"--c",label="cos(x)")
plt.legend()  # shows information about which line belongs to which line
```

```python
# • plt.xlabel → ax.set_xlabel
# • plt.ylabel → ax.set_ylabel
# • plt.xlim → ax.set_xlim
# • plt.ylim → ax.set_ylim
# • plt.title → ax.set_title
```

```python
ax=plt.axes()
ax.plot(x,np.sin(x),".r")
ax.set(xlabel="X value",ylabel="sin(x)",xlim=(0,10),ylim=(-1,1.5),title="Sine Curve");
```
